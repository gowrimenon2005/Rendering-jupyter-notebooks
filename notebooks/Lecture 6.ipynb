## Unified Range Equation

In the previous lecture ([Lecture 5](Lecture%205.ipynb)) the equations for maximum range were derived for both jet and propeller aircrafts. The equations are presented below for your convenience. 

$$ R_{\text{max - jet}} = \frac{V}{c_T} \frac{C_L}{C_D} \ln{\frac{W_{ini}}{W_{fin}}} $$

$$ R_{\text{max - prop}} = \frac{\eta}{c_P} \frac{C_L}{C_D} \ln{\frac{W_{ini}}{W_{fin}}} $$

Both of these equations are mainly composed of the same variables and have a similar structure. In fact except for the first term in the right-hand side of the product, the equations are identical. In order words, if it were possible to express $\frac{V}{c_T}$ and $\frac{\eta}{c_P}$ in some 'unified' manner the equations would be identical. In such a case, it is reasonable to wonder whether there exists a way to combine these equations into one. This idea is investigated through the notion of the unified range equation. To do so, let us first introduce the parameter of total propulsive efficiency. This is denoted by:

$$ \eta_{tot} = \frac{P_a}{Q} $$

Where, $Q$ is the thermal power in the system. Hence it can be written as:

$$ Q = FF \frac{H}{g} $$

Where, $FF$ is the fuel flow, and $H$ is the energy density (the energy per kilogram of fuel). Substituting this relation into the equation for total propulsive efficiency, one obtains the following:

$$ \eta_{tot} = \frac{P_a}{Q} = \frac{g}{H \cdot FF} P_a $$
 
At this stage, it is essential to distinguish between jet and propeller aircrafts. For jet aircrafts, the following can be stated (noting that the power available is the product of velocity and thrust):

$$ \eta_{tot} = \frac{g}{H \cdot FF} TV = \frac{g}{H} \frac{V}{c_T} $$

In the case of propeller aircrafts, the power available can be expressed in terms of the brake-shaft power i.e., $P_a = \eta P_{sh}$, thereby obtaining the following equation:

$$ \eta_{tot} = \frac{g}{H \cdot FF} \eta P_{sh} = \frac{g}{H} \frac{\eta}{c_P} $$

Under this decomposition, it becomes clear that both $\frac{V}{c_T}$ and $\frac{\eta}{c_P}$ can be expressed in a 'unified' manner. To see this clearly, observe: 

$$ \frac{V}{c_T} = \frac{ \eta_{tot} H}{g} \\ \frac{\eta}{c_P} = \frac{\eta_{tot} H}{g} $$

Hence, the range equation can be rewritten to state:

$$ R = \frac{H}{g} \eta_{tot} \frac{C_L}{C_D} \ln{ \frac{W_{ini}} {W_{fin}}} $$

This is the unified range equation! This equation is independent of the propulsive system type! Upon further consideration, this formula is quite elegant. It concisely captures the dependence of fuel quality, propulsive efficiency, aerodynamic efficiency and weight efficiency on range. Quite beautiful, is it not?

FURTHER CONSIDERATION OF WEIGHT!

$$ \frac{W_{ini}}{W_{fin}} = \frac{W_{OE}+W_{PL}+W_{f_{reserve}}+W_{f_{used}}+W_{f_{unused}}}{W_{OE}+W_{PL}+W_{f_{reserve}}+W_{f_{unused}}} $$

$$ \frac{W_{ini}}{W_{fin}} = 1+ \frac{W_{f_{used}}}{W_{OE}+W_{PL}+W_{f_{reserve}}+W_{f_{unused}}} $$



## Economics

A very important consideration under any operational procedure is that of profitability. Specifically when investigating cruise economics one diagram becomes particularly detrimental, the payload-range diagram. Although this diagram has been introduced to you in previous courses, it is presented again here for completeness. A general payload diagram has the following form: 

![IMAGE](Images\lec6-payload-range-gen.png)

Here a few points are to be noted: 
- The operational empty weight is constant for all ranges
- The reserve fuel is also constant for all ranges 
- The initial slope in the payload range diagram (starting from maximum zero fuel weight to maximum take-off weight) gives the take-off weight at the maximum payload that can be carried. 
- The design range corresponds to the maximum stage length with maximum payload
- To increase range after the design range, payload must be replaced by fuel. Doing so one is able to obtain the ultimate range. Note that to achieve the ultimate range the aircraft must fly with zero payload and reserve fuel unconsumed.

Some other parameters that are useful for comparison, include the block speed, block speed and transport productivity. Block speed is defined as the time it takes from gate to gate. It is computed using the following equation:

$$ t_{block} = \frac{R}{V_{cr}} + \Delta t (R) $$

Block speed is the average speed during the block time and hence can be calculated by:

$$ V_{block} = \frac{R}{t_{block}} = \frac{R}{\frac{R}{V_cr}+\Delta t(R)} $$

Transport productivity is the amount of payload that can be transported over a certain distance in a certain time. This follows logically from the equations stated above:

$$ TP = W_{PL} \cdot \frac{R}{t_{block}} = W_{PL} \cdot V_{block} $$

Furthermore, as one could imagine the costs of operating an aircraft come in several different forms and magnitudes. Loosely such costs can be categorised into either direct operating costs or indirect operating costs. Direct operating costs are proportional to the range covered and the number of flights. Examples of direct operating costs could include: fuel, salary, maintenance, loans, insurance, etc. Alternatively indirect operating costs could be texes, passenger service expenses, general administrations, etc. 

## Transonic efficiency 

To conclude our analysis of cruise operations, let us consider transonic, aerodynamic efficiency. 

$$ \frac{V}{FF} = \frac{a}{c_T} M \frac{C_L}{C_D} \frac{1}{W} $$

Similar to before, we aim to maximise the ratio of velocity to fuel flow. Hence, in this case, $M \frac{C_L}{C_D}$ must be maximised. Under transonic conditions, this is not a trivial condition. This is because, both the lift and drag coefficient are dependent on Mach number. If this were not the case, we arrive to the optimal conditions derived in the fifth notebook. 

To begin the analysis, we begin with the following assumptions:
- $$ C_L = C_L(M) $$
- $$C_D = C_D(M) $$
- $$c_T = constant $$
Furthermore, as consider the following form for the drag polar: 
$$ C_D = C_{D0}(M) + k C_L(M)^2 $$

Using these assumptions and for some supersonic data, one can plot the lift coefficient against the drag coefficient for several Mach numbers. This may seem arbitrary but remember, our aim is to maximise $M \frac{C_L}{C_D}$. Hence, realise for each drag polar curve at some point the greatest value of $M \frac{C_L}{C_D}$ can be noted. This is done in the plot below. 

```python
# from function_file import * 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import *
# from data_file import *
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
# import scipy 
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")

def get_supersonic_data(mach):
    dat = pd.read_csv(f'./supersonic_data/cl_cd_mach_{mach}..csv')
    cl = dat['cl']
    cd = dat['cd']
    # quadratic fit
    p = np.polyfit(cl, cd, 2)
    return p

def transonic_cl_cd():
    sns.set_style('whitegrid')
    # Load factor limits: aerodynamic, propulsive, and structural
    mach = [0.45, 0.65, 0.79, 0.81, 0.83, 0.85, 0.9, 0.92]
    for i in range(len(mach)):
        p_vals = get_supersonic_data(mach[i])
        cl = np.linspace(0, 0.7, 100)
        cd = p_vals[0] * cl**2 + p_vals[1] * cl + p_vals[2]
        plt.plot(cl, cd, label=f'Mach {mach[i]}')

    plt.xlabel('CL')
    plt.ylabel('CD')
    plt.legend()
    plt.show()

transonic_cl_cd()
```

This is helpful, but still quite ambiguous as it requires one to infer the effects of Mach number. To clarify let us plot lift coefficient against the ratio of lift to drag coefficient for each fixed Mach number. Furthermore, for each fixed Mach number calculate the largest ratio of lift to drag obtained. This results in the following plot. 

```python
def transonic_cl_clcd():
    sns.set_style('whitegrid')
    # Load factor limits: aerodynamic, propulsive, and structural
    mach = [0.45, 0.65, 0.79, 0.81, 0.83, 0.85, 0.9, 0.92]
    for i in range(len(mach)):
        p_vals = get_supersonic_data(mach[i])
        cl = np.linspace(0, 0.7, 100)
        cd = p_vals[0] * cl**2 + p_vals[1] * cl + p_vals[2]
        plt.plot(cl, cl/cd, label=f'Mach {mach[i]}')
        # plot max cl/cd
        max_clcd = cl[np.argmax(cl/cd)]
        plt.plot(max_clcd, max(cl/cd), 'bo', markersize=5, markerfacecolor='b')
    plt.xlabel('CL')
    plt.ylabel('CL/CD')
    plt.legend()
    plt.show()

transonic_cl_clcd()
```

From this plot it becomes clear that as the Mach number increases, the maximum lift to drag ratio decreases. This can be illustrated a bit more clearly in a carpet plot. Please note, here the Mach number is plotted against the lift coefficient for each contour of $M \frac{C_L}{C_D}$. The global maximum occurs near the center. 

```python
def transonic_m_cl():
    sns.set_style('whitegrid')
    # Load factor limits: aerodynamic, propulsive, and structural
    mach = [0.45, 0.65, 0.79, 0.81, 0.83, 0.85, 0.9, 0.92]
    cl_vals = np.linspace(0, 0.7, 100)
    mach_val = np.array(mach)
    cl, mach_vals = np.meshgrid(cl_vals, mach_val)
    contour = np.zeros((len(mach_vals), len(cl_vals)))
    
    for i in range(len(mach)):
        p_vals = get_supersonic_data(mach[i])
        cd = p_vals[0] * cl_vals**2 + p_vals[1] * cl_vals + p_vals[2]
        val = cl_vals / cd * mach[i]
        # print(len(val))
        contour[i, :] = val
    const_cl_m1 = cl_vals * mach_vals**2
    contours = plt.contourf(mach_vals, cl, contour, levels=15, cmap='viridis')
    plt.contour(mach_vals, cl, contour, levels=15, colors='k', linestyles='dashed')
 
    plt.xlabel('Mach')
    plt.ylabel('CL')
    plt.colorbar(contours, label='CL/CD * Mach')
    plt.show()

transonic_m_cl()
```

It is not possible to fly everywhere in the transonic efficiency carpet, as the following relation must still hold: 

$$ C_L = \frac{2}{\rho} \frac{w}{S} \frac{1}{V^2} = \frac{2}{\rho} \frac{w}{S} \frac{1}{a^2} \frac{1}{M^2} $$

Now, if we define the trim lines as lines of $C_L M^2$, then along the trim lines $\frac{W}{\rho a^2}$ is constant. 

Therefore, it must be noted that for a given weight and altitude, the optimal Mach and lift coefficients are found by the tangency point between the current trim and the highest contour line of transonic efficiency. The center of the carpet plot is ideally the global optimum. Thus, if weight decreases and altitude increases, $C_L$ and Mach number must be changed together such that $C_L M^2$ remains constant. 

Now with the plot below you can investigate the effects of weight and altitude. 

```python
from ipywidgets import *

weight = widgets.FloatSlider(min=10000, max=50000, step=1000, value=5000, description='weight')
# rho = widgets.FloatSlider(min=0.3639, max=1.225, step=0.1, value=1.0, description='air density')
altit = widgets.FloatSlider(min=0, max=10000, step=100, value=0, description='altitude')

weight.style = {'description_width': 'initial'}
# rho.style = {'description_width': 'initial'}
altit.style = {'description_width': 'initial'}
display(weight)
# display(rho)
display(altit)

# transonic_m_cl_alt(weigth=weight, rho=rho)

button = widgets.Button(description="Plot")

def transonic_m_cl_alt(weight, altit):
    sns.set_style('whitegrid')
    temp = 288.15 - 0.0065 * altit
    rho = 1.225 * (temp/288.15)**(9.81/(0.0065*287))
    # Load factor limits: aerodynamic, propulsive, and structural
    mach = [0.45, 0.65, 0.79, 0.81, 0.83, 0.85, 0.9, 0.92]
    cl_vals = np.linspace(0, 0.85, 100)
    mach_val = np.array(mach)
    cl, mach_vals = np.meshgrid(cl_vals, mach_val)
    contour = np.zeros((len(mach_vals), len(cl_vals)))
    
    for i in range(len(mach)):
        p_vals = get_supersonic_data(mach[i])
        cd = p_vals[0] * cl_vals**2 + p_vals[1] * cl_vals + p_vals[2]
        val = cl_vals / cd * mach[i]
        # print(len(val))
        contour[i, :] = val
    
    # trim lines 
    alt = 288.15 / 0.0065 * (1 - (rho/1.225)**(-9.81/(0.0065*287)))
    # print(alt)
    trim_cl = weight / rho * 1/((287 * 1.4 * (288.15-0.0065*alt))) * np.ones_like(mach_val)
    trim_cl = trim_cl / mach_val**2
    contours = plt.contourf(mach_vals, cl, contour, levels=15, cmap='viridis')
    plt.contour(mach_vals, cl, contour, levels=15, colors='k', linestyles='dashed')
    plt.plot(mach_val, trim_cl, 'r--', label='Trim line')
    plt.xlabel('Mach')
    plt.ylabel('CL')
    plt.colorbar(contours, label='CL/CD * Mach')
    plt.ylim(0, 0.85)
    plt.show()


output = interactive_output(transonic_m_cl_alt, {'weight': weight, 'altit': altit})

# display(button)

def on_button_click(b):
    output = interactive_output(transonic_m_cl_alt, {'weight': weight, 'altit': altit})
    output.clear_output()
    with output:
        transonic_m_cl_alt(weight=weight.value, rho=altit.value)

button.on_click(on_button_click)

display(output)
```

From this plot, when changing the inputs you might have noticed two important characteristics:
1. As weight decreases under constant Mach number, the trim line moves left/down. The implication of this result is that $C_L$ must decrease and hence the flight point moves down. If you do not see why, recall: 
$$ C_L M^2 = const \left( \frac{W}{\rho a^2} \right) $$
2. As weight decreases while altitude is constant, the trim line moves left/down. If $C_L$ remains constant, then Mach number must decrease and hence the flight point moves to the left. 



## Realistic considerations

The analysis we have considered so far dealt with closed form solutions. 

Although closed form solutions are useful for initial computations, for more detailed analysis numerical methods are used. To illustrate the idea behind numerical methods let us consider the range of realistic turbofan aircraft. From the considerations seen so far it is evident that to maximum transonic efficiency, one must fly at the points which yield the highest value of $M \frac{C_L}{C_D}$. Unfortunately, this point changes with altitude. In addition, the specific fuel consumption of the aircraft also varies with altitude. At this stage finding a closed form graphical solution is quite complicated and hence let us resort to numerical methods. One method of solving this problem is to create a mesh of points within a reasonable flight envelope, such as in the figure below:

![text](Images\lec6-mesh.png)

Note that each of the points in this mesh correspond to a particular specific altitude and specific speed. After which the following methodology can be used: 
1. For each point, compute $C_L$ and $C_D$ 
2. Using the determined value of $C_D$, the total drag can be determined which in cruising flight is equal to the thrust of the aircraft. 
3. Assuming the engine data is known and hence the value of $c_T$, one can compute the fuel flow. 
4. Calculate the associated $\frac{V}{FF}$ 
Repeat this procedure for all points in the mesh. The maximum value of $\frac{V}{FF}$ is the optimum flight condition.

Usually in practice, maximum range is sacrificed for greater productivity.
