## Flight Conditions for Maximum Range
Prior to deriving the equations of range, let us first develop some intuition for the conditions under which maximum range can be obtained. 

<!-- One could correctly imagine this condition would be different based on the propulsive system used. Hence, initially the flight conditions of maximum range for propeller aircrafts will be investigated. After which, the same analysis will be done for jet aircrafts.  -->

Range refers to the horizontal distance covered by an aircraft in cruising flight. Maximum range thus refers to the maximal horizontal distance an aircraft could fly and hence is limited by the aircraft's fuel capacity. From basic kinematics it can be noted that range is the definite integral of velocity with respect to time. The total time spanned can be calculated by considering the fuel flow i.e., fuel consumption rate (as fuel capacity is the limiting factor for maximum range). Let us denote fuel flow as:

$$ FF = \frac{dW_f}{dt} $$

Now it is critical to realise that the change in fuel weight is equal to the negative of change in aircraft weight. This is because burning fuel decreases the weight of the aircraft. Hence: 

$$ dW_f = -dW \Rightarrow FF = -\frac{dW}{dt}$$

Combining these qualitative realisations, the following equation can be written:

$$ R = \int_{0}^{R} dx = \int_{0}^{t_{R}}V dt = \int_{W_{initial}}^{W_{final}} -\frac{V}{FF} dW$$
where recalling the definition of fuel flow leads to:
$$ FF = -\frac{dW}{dt} \Rightarrow R_{max} \Leftrightarrow \left( \frac{V}{FF} \right)_{max} $$

In other words, to maximise the range obtained one must maximise the ratio of velocity to fuel flow. 
To proceed with any further analysis, one must determine $FF$. The determination of fuel flow is dependent on the propulsive type of the aircraft. Hence further computations must be analysed seperately for propeller and jet aircrafts. First jet aircrafts will be considered. 

### Simplified Jet Aircraft
For jet aircrafts fuel flow is defined by the thrust, $T$ and specific fuel consumption, $c_T$: 

$$ FF = c_T T$$ 

Hence, substituting this into the condition for maximum range, one obtains:
$$ FF = c_T T \rightarrow \left( \frac{V}{FF} \right)_{max} \Leftrightarrow \left( \frac{V}{T} \right)_{max} \Leftrightarrow \left( \frac{V}{D} \right)_{max} \Leftrightarrow \left( \frac{D}{V} \right)_{min} $$

In other words, for jet aircrafts the condition of maximum range corresponds to the flight condition where the ratio of drag to velocity is minimised. Similar to before let us consider the equation of drag assuming a two term drag polar:

$$ D = \frac{1}{2} \rho V^2 S (C_{D_{0}}+ k C_{L}^2) = aV + \frac{b}{V^3} $$

where, a and b are the same as before but denoted below for ease of readability:

$$ a = \frac{1}{2} \rho S C_{D_{0}} \\ b = \frac{2}{\rho}\frac{W^2}{S} k $$

Takign the partial derivative of $\frac{D}{V}$ with respect to velocity yields the following:

$$ \frac{\partial (D / V)}{\partial V} = a - \frac{3b}{V^4} = 0 \Rightarrow aV^2 = 3\frac{b}{V^2} $$

Hence, to minimise the ratio of drag to velocity, the drag coefficient must be:
$$ C_{L_{(D /V)_{min}}} = \sqrt{\frac{1}{3} \pi A e C_{D_{0}}} $$

And, the corresponding lift coefficient is:

$$ C_{D_{(D/ V)_{min}}} = \frac{4}{3}C_{D_0} = \frac{3C_L^2}{\pi A e} $$

Here, maximum range is dependent on both velocity and aerodynamics. As altitude increases the velocity required to maintain the optimal aerodyanmic condition changes favourably. Hence, the greatest range is obtained at the highest altitude. This altitude is determined by propulsive and operational limits. As discussed briefly in the first lecture, there is a maximum Mach number that may not be exceeded for various reasons such as aero-interactions, the addition of aerodynamic loads, lack of controllability. To summarize the optimal flying strategy for jet aircraft o maximum range is to fly as high as possible until operation or propulsive limits are encountered. 

### Simplified Propeller Aircraft

For propeller-driven aircrafts, fuel flow is defined by the brake-shaft power, $P_{sh}$, and the specific fuel consumption, $c_P$. Recall the brake-shaft power is defined as:

$$ P_{sh} = \frac{P_a}{\eta_j} $$

where, $\eta_j$ is the propulsive efficiency. Hence substituting these relations into the condition for maximum range, one obtains: 

$$ FF = c_P P_{sh} \rightarrow \left( \frac{V}{FF} \right)_{max} \Leftrightarrow \left( \frac{V}{P_{sh}} \right)_{max} \Leftrightarrow \left( \frac{V}{TV} \right)_{max} \Leftrightarrow D_{min} $$

In other words, the flight condition under which maximum range is obtained is equivalent to the flight condition for minimum drag. Drag is defined by the following equation, where a two term drag polar is used:

$$ D = \frac{1}{2} \rho V^2 S (C_{D_{0}}+ k C_{L}^2) $$

Furthermore, due to the assumption of level flight, weight is equal to lift and hence the lift coefficient can be written as:

$$ C_L = \frac{2}{\rho} \frac{W}{S} \frac{1}{V^2} $$

Substituting this relation into the equation of drag, one obtains an equation of the following form:
$$ D = a V^2 + \frac{b}{V^2} $$

where, 

$$ a = \frac{1}{2} \rho S C_{D_{0}} \\ b = \frac{2}{\rho} \frac{W^2}{S} k $$

This representation is useful as to determine the condition for minimum drag, the partial determine of drag with respect to velocity must be determined. Performing the computations the following is obtained:

$$ \frac{\partial D}{\partial V} = 2aV - \frac{2b}{V^3} = 0 \Rightarrow aV^2 = \frac{b}{V^2}$$

Substituting this into the two term drag polar assumed, the drag coefficient for minimum drag is:
$$ C_{D_{D_{min}}} = 2C_{D_{0}} = 2\frac{C_L^2}{\pi A e}$$

The corresponding lift coefficient for minimum drag is:
$$ C_{L_{D_{min}}} = \sqrt{\pi A e C_{D_{0}}} $$

Furthermore, this brings an important observation to light, because maximum range depends only on pure aerodynamics (which is independent of altitude) maximum range does not depend on altitude. 

## Range of Simplified Jet 

Having seen the conditions for maximum range, we are almost at a position derive the expression of maximum range. The effect of weight must still be discussed. So far we have assumed weight is constant, however during cruise is when most fuel is burnt hence it is not reasonable to assume weight remains constant. If weight decreases the optimal lift and drag coefficient decrease and hence optimal velocity decreases too. Thus, it appears the optimal strategy is to slow down at constant altitude...

This is not true! If the weight decreases there exists excess thrust and hence we can climb to a higher altitude (which is always favourable)! To understand why this is the case practically, consider the pilot flys at a constant aerodynamic condition and throttle setting. At this situation the aircraft loses weight and gradually gains altitude. Because thrust decreases exactly proportionally to air density, the velocity remains constant unless subjected to other pilot inputs. This means the optimal flight strategy for jet aircrafts is a continuous cruise climb at the optimum aerodynamic conditions. 

$$ R = \int_{W_{initial}}^{W_{final}} -\frac{V}{FF} dW = \int_{W_{final}}^{W_{initial}} \frac{V}{c_T D} dW = \int_{W_{final}}^{W_{initial}} \frac{V}{c_T} \frac{C_L}{C_D} \frac{1}{W} dW$$

Here, the lift and drag coefficient are constant under the flight strategy considered and thus the integral becomes:

$$ R = \frac{V}{c_T} \frac{C_L}{C_D} \ln{ \frac{W_{initial}} {W_{final}}} $$

We have previously established the optimal flgiht strategy is a continuous cruise climb. However, in practice this is not done. Air Traffic Control requires pilots to maintain fixed altitudes during cruise, as a result the continuous cruise climb becomes a stepped cruise climb. This is therefore an operational approximation of the optimal strategy. 


## Range of Simplified Propeller 

Unlike jet aircrafts, the maximum range of propeller aircrafts is independent of velocity and altitude. Hence, the derivation can be completed without any further considerations. This can be done by using the initial formula used for all analysis. That is using the kinematics of the problem, the maximum range can be written as follows:

$$ R = \int_{W_{initial}}^{W_{final}} -\frac{V}{FF} dW = \int_{W_{final}}^{W_{initial}} \frac{\eta}{c_P} \frac{V}{P_a} dW = \int_{W_{final}}^{W_{initial}} \frac{\eta}{c_P} \frac{C_L}{C_D} \frac{1}{W} dW$$

From the understanding of the required flight conditions, it becomes clear that both lift and drag coefficient should remain constant to obtain maximum range. Hence the integral resolves into:

$$ R = \frac{\eta}{c_P} \frac{C_L}{C_D} \ln{ \frac{W_{initial}} {W_{final}}} $$

## Optimising for Endurance 
Like range, another important cruise parameter is endurance. This refers to the amount of time spend in cruising flight. As a result endurance is closely rated to the specific fuel consumption of the aircraft. Mathematically, endurance can be written as:

$$ E = \int_{t_{initial}}^{t_{final}} dt = \int_{W_{initial}}^{W_{final}} -\frac{dW}{FF} = \int_{W_{final}}^{W_{initial}} \frac{dW}{FF} $$

From this formula it becomes clear that a distinction must be made for jet and propeller aircrafts, as the specific fuel consumption parameter is dependent on the propulsive system. 

### Jet aircrafts
Following a similar analysis procedure to range, the condition for maximising endurance can be found by finding the condition for minimising the specific fuel consumption. Thus,

$$ FF = c_T T \rightarrow (FF)_{min} \Leftrightarrow (T)_{min} \Leftrightarrow (D)_{min} $$

From here it becomes apparent that for maximising endurance one must fly at the condition for minimum drag. This is the same condition that maximises range for a propeller aircraft and thus the optimal aerodynamic condition is:
$$ C_{D_{D_{min}}} = 2C_{D_{0}} = 2\frac{C_L^2}{\pi A e}$$
and, 
$$ C_{L_{D_{min}}} = \sqrt{\pi A e C_{D_{0}}} $$

Based on this the endurance can be computed as follows:
$$ E = \int_{W_{final}}^{W_{initial}} \frac{1}{c_T}\frac{C_L}{C_D}\frac{dW}{W} = \frac{1}{c_T}\frac{C_L}{C_D}\ln \frac{W_{initial}}{W_{final}} $$

### Propeller aircrafts
For propeller aircrafts, minimising the specific fuel consumption results in the following train of thought:
$$ FF = c_P P_{sh} \rightarrow (FF)_{min} \Leftrightarrow ({P_{sh}})_{min} \Leftrightarrow (TV)_{min} \Leftrightarrow (DV)_{min} = \Leftrightarrow \left( \frac{C_D^2}{C_L^3}\right)_{min} = \Leftrightarrow \left( \frac{C_L^3}{C_D^2}\right)_{max}$$

If you do not see why the last two implications hold, consider the following:

$$ (DV)_{min} = \left(\frac{C_D}{C_L} W \cdot \sqrt{\frac{W}{S} \frac{2}{\rho} \frac{1}{CL}} \right)_{min} = \left( W \sqrt{\frac{W}{S} \frac{2}{\rho} \frac{C_D^2}{C_L^3}} \right)_{min} $$

Here all parameters are constant except for the aerodyanmic conditions. Of course weight is time dependent but that is 'taken care' of in the integral (as we integrate over time). 

Hence, the expression for maximum endurance becomes:

<!-- $$ E = \int_{0}^{T} dt = \int_{W_{initial}}^{W_{final}} - \frac{dW}{FF} = \int_{W_{final}}^{W_{initial}} \frac{dW}{FF} $$ -->

$$ E = \int_{W_{final}}^{W_{initial}} \frac{\eta_j}{c_P} \frac{dW}{W \sqrt {\frac{W}{S}\frac{2}{\rho}\frac{C_D^2}{C_L^3}}} $$

$$ E = \frac{\eta_j}{c_P}\sqrt{\frac{C_L^3/C_D^2}{\frac{1}{S}\frac{2}{\rho}}} \int_{W_{final}}^{W_{initial}} \frac{dW}{W\sqrt{W}} = \frac{\eta_j}{c_P}\sqrt{\frac{C_L^3/C_D^2}{\frac{1}{S}\frac{2}{\rho}}} \left[ \frac{2}{\sqrt{W_{final}}} - \frac{2}{\sqrt{W_{initial}}} \right] $$




## Effects of Wind

One of the assumptions made earlier on in this lecture was that there is no wind. This is not fully realistic and hence let us pay some attention to the effects of wind on range. Wind patterns vary greatly, but in this section only constant head or tailwind directed along the flight path will be considered. The concept of relative velocities was previously introduced and using that notation, in (quasi-) level flight ground velocity can be written as: 

$$\{\vec{V_{ac}}\}_{ground} = \{\vec{V_{ac}}\}_{air} + \{\vec{V_{air}}\}_{ground} = \vec{V}+ \vec{V}_{wind} $$

Here, (similar to before), tailwind is taken as positive, and headwind is negative. 

To proceed with further analysis is essential to realise wind does not influence endurance, rather endurance is simply based on the fuel consumption rate. However range is dependent on both velocity with respect to the ground and the fuel consumption rate. As a result range is greatly infleunced by wind. To understand how one can simply substitute the relation for the velocity of the aircraft with respect to the ground based on the wind speed and airspeed of the aircraft. Doing so the range equation becomes: 

$$ R = \int_{W_{initial}}^{W_{final}} -\frac{V_g}{FF} dW = \int_{W_{final}}^{W_{initial}} \left( \frac{V}{FF} + \frac{V_{wind}}{FF} \right) dW = \int_{W_{final}}^{W_{initial}} \frac{V}{FF} + V_w \int_{W_{final}}^{W_{initial}} \frac{dW}{FF} $$

The final term in the equation above is the endurance of an aircraft as it computes the total time the aircraft can remain in the air. Hence, under the presence of constant tail/head-wind, range is dependent on endurance. As a result, range in the presence of wind can be written in terms of range under no wind and endurance in the following manner:
 
$$ R = R_{V_{w}=0} + V_w \cdot En $$

Here it becomes clear tailwind increases the cruise range and headwind decreases the cruise range.
