## Take-off Maneuver
### V-speeds and flight manual
Take-off is defined as the maneuver by which the aircraft is accelerated from rest to the climb-out speed. The distance required to do so is the take-off distance and is composed of two parts: the ground run distance and airborne distance. The details of the maneuver are elaborated on in the next section [Take-off maneuver](#take-off-maneuver). Prior to that analysis, it is useful to understand the practical nature of the maneuver. To illustrate the maneuver is shown below:

![text](\Images\lec4-vspeed.png){ width=400px }

The screen height is an important measure, and is defined by regulations. It is 10.7m for civil transportation and 15.2m for light propellers and military aircraft. During the maneuver the main performance metrics are take-off distance (this is most important for airlines), take-off mass (this is critical for designers) and take-off time (this is essential for air traffic management, ATM). Furthermore, there are three main operative scenarios: nominal take-off, aborted take-off and one engine out (OEO).

Referring back to the figure above, you might have noticed the several speeds indicated. These speeds are of significance as they exist to ensure adequate safety and handling. Although, this may seem rather vague at first, it purpose of these speeds should become clear rather soon. Below these speeds are listed in the order they are usually encountered during the maneuver. In addition, a figure of these speeds in the order they are encountered is given.

![text](\Images\lec4-all-speeds.png)

1. $V_{ms}$: this is the minimum stall speed in take-off configuration. 

2. $V_{mc}$: this is the minimum speed above which it is possible to maintain adequate control of the aircraft after one engine out (OEO). 

3. $V_{mcg}$: this is the minimum speed above which it is possible to maintain adequate control of the aircraft on the ground (without nose wheel) 

4. $V_{ef}$: this is maximum speed at which one engine might fail (aircraft manufacturer, no pilot delay) 

5. $V_{1}$: decision speed. This speed is chosen such that if an engine failure is recognised at this speed, the pilot can decide whether to abort the take-off or continue the take-off with one engine out. The decision speed is determined by computing both the accelerate-stop and accelerate-climb distance as a function of $V_{ef}$. The former distance refers to the length required to accelerate to $V_{ef}$ and thereafter stop. Whereas the latter refers to the distance required to accelerate to $V_{ef}$ and then continue take-off with OEO. The speed at which both of these distances are equal is the decision speed. The corresponding take-off distance is the balanced field length. This is shown below:

![text](\Images\lec4-bfl.png)

To give a practical consideration: the required field length is the largest between the balanced field length and nominal take-off distance plus 15%, i.e., 

$$\text{Required field length} = \max{(\text{Balanced field length} \vee s_{to} + 15\%)}$$

6. $V_{r}$: this is the initiate rotation speed. It must be chosen such that: $V_{lof} \geq 1.1V_{mu}$ with all engines operating or $V_{lof} \geq 1.1V_{mu}$ with OEO. Furthermore, this speed must ensure that $V_{2}$ speed is reached at screen height.  

7. $V_{mu}$: minimum lift-off or unstick speed. At this calibrated speed the aircraft can safely lift-off the ground.. 

8. $V_{lof}$: lift-off speed in nominal take-off. The actual lift-off speed depends on the rotation speed.

9. $V_{2}$: this is the minimum speed at which safe climb-out is ensured even with the critical engine inoperative and the live engines operating at full take-off thrust. 
 
Lastly, please note, V-speeds and other performance speeds are determined by the manufacturer. Tabular values of these values are available depending on the aircraft configuration and environment conditions. In this context, aircraft configuration relates to the weight, maximum lift coefficient, etc. Environment conditions, are dependent on the altitude, temperature, humidity, dryness/wetness, etc. All of this information is written in the Flight Manual.


## Take-off Performance

Having seen the take-off maneuver conceptually, it is no surprise that it is a dynamic, and complex process. To simplify the analysis of this process a number of assumptions will be made. These assumptions and their implications are critical to the relations derived hereafter. 
1. The ground effect has been neglected: in reality however the ground effect will result in increased production of lift and decreased production of drag. This is favourable for take-off conditions.
2. Wind has been neglected: the analysis performed here will be done for still air. In the presence of wind the derivations differ as the speed of the aircraft with respect to the ground will change. The difference will be considered at the end of the lecture 
3. The runway has no slope: this assumption directly affects the equations of motion derived. However, once the procedure is fully understood you should be able to replicate the derivations for a sloped runway.
4. The following derivations also make use of semi-empirical methods for first order sizing. These values have been inferred from historical data. Please pay attention to this.

The maneuver consists of two phases: the ground run and the airborne phase. The airborne phase can be further divided into: the transition region and the climb-out region. The total distance covered by the maneuver is the sum of the horizontal distances spanned by each of the individual regions. Due to the linearity of distance, the individuals phases can be considered separately for ease of analysis. The entire maneuver is visualised below:

![image.png](\Images\lec4-takeoff.png)

Firstly, let us consider the ground run. During ground run, the forces acting on the aircraft are depicted below.

![text](\Images\lec4-fbd-ground-run-to.png)

From the diagram it is evident that the weight of the aircraft is in equilibrium with the lift and reaction forces from the ground surfaces on the wheels. Furthermore, thrust generated by the aircraft is opposed by the drag of the aircraft and the rolling friction between the wheels and ground surface. This can be written mathematical in the equations of motion, i.e.,

$$ L + N - W = 0 \\ T - D - D_g = m \dot{V} $$

The rolling friciton between the wheels and ground surface is a fraction of the reaction forces the ground surfaces on the wheels. Where, the fraction is given by the coefficient of rolling friction, $\mu_r$. Hence, $D_g$ can be represented as: 

$$ D_g = \mu N = \mu (W-L) $$

Based on this, the horizontal acceleration can be written as: 

$$ \dot{V} = \frac{g}{W}[T-D-\mu(W-L)] $$

This horizontal acceleration, can be denoted as the average acceleration and can be rearranged to state:

$$ \bar{a} = \dot{V} = g \left[ \frac{T}{W} - \frac{D}{W} - \mu \left( 1 - \frac{L}{W} \right) \right] $$

Furthermore, recall from basic kinematics, acceleration is the second derivative of displacement and hence the distance covered can be computed by:

$$ s_g = \int_0^{x_{lof}} dx = \int_0^{t_{lof}} V dt = \int_0^{V_{lof}} \frac{V}{a} dV = \frac{1}{\bar{a}} \int_0^{V_{lof}} VdV = \frac{V_{lof}^2}{2\bar{a}} $$

As an approximation for first sizing, the lift-off velocity can be regarded as a "multiple" of the minimum, stall, velocity. Please note, the actual lift-off velocity is dependent on the rotation speed. 

$$ V_{lof} = 1.05 V_{min} = 1.05 \sqrt{\frac{2}{\rho} \frac{W}{S} \frac{1}{C_{L_{max, to}}}} $$

Secondly, let us consider the transition region of the airborne phase. The running assumption through this analysis will be that the velocity is constant and is equal to the lift-off velocity. The diagram of the trajectory is shown below:

![alt text](Images\lec4-takeoff-traj.png)

From the geometry of the trajectory, the following becomes apparent.

$$ L-W = \frac{w}{g}{V^2}{R} \Rightarrow n-1 = \frac{1}{g}{V^2}{R} $$

Again, here a first order sizing approximation will be made, as the load factor will be assumed to be 1.15. Under such an assumption, the following can be written:

$$ R = \frac{V_{lof}^2}{0.15g} \Rightarrow s_t = R \sin \gamma_{cl} $$ 

$$ s_t = \frac{V_{lof}^2}{0.15g}\sin{\gamma_{cl}} $$

Lastly, the climb-out phase can be analysed. This is visualised below:

![alt text](Images\lec4-takeoff-traj2.png)

Based on the geometry of the region, using simple Pythagorean identities, the following can be said:

$$ \Delta h_{cl} = h_{scr} - (R - R\cos \gamma_{cl}) $$

$$ s_{cl} = \frac{h_{scr} - R(1-\cos \gamma_{cl})}{\tan \gamma_{cl}} $$ 

$$ s_{cl} = \frac{h_{scr} - \frac{V_{lof}^2}{0.15g}(1 - \cos \gamma_{cl})}{\tan \gamma_{cl}} $$

Having defined expression for each of the different regions, the total distance can be found by:

$$ s_{total} = s_{g} + s_{t} + s_{cl} $$

From this equation, the following relations can be inferred. If lift-off velocity decreases, or the climb angle increases, or weight decreases, or thrust increases then the take-off distance decreases. Make sure you can justify these dependencies!

## Factors affecting Take-off Performance 

As stated previously, take-off is a dynamic and complex process. It is highly dependent on several factors such as aircraft design, flight parameters and regulations. Here a couple of the most important factors affects take-off are considered. Please note this list is not extensive. 
- **Engine Thrust**: The production of thrust is quintessential for performing the take-off maneuever. <br><br>

- **Contact**: The ground contact surface is essential for determining the coefficient of dynamic friction experienced by the tyres. This is dependent on the material type of the tyres, the weight of the aircraft, the pressure of the tyres, the material of the runway and the dryness/wetness of the runway. <br><br>

- **Runway slope**: The presence of a positive runway slope (positive defined in the counter-clockwise direction) adversely impacts the take-off distance. This can be seen by constructing a free body diagram. Intuitively it can be understood by the following: in the presence of a positive runway slope, the ground acceleration of the aircraft is lowered because the slope introduces an additional weight term opposing the motion of the aircraft. <br><br>

- **Lift-off speed**: <br><br>

- **High-lift devices**: High-lift devices can be critical to increase the maximum lift coefficient during take-off, thereby leading to a higher lift. However, the addition of these devices often comes at the cost of increase weight and increased cost. <br><br>

- **Drag**: To reduce take-off distance, drag must be decreased. However, drag and lift are both dependent on the same parameters, thus one aims to decrease one whilst maximising the other (a conflict of interest so to speak). <br><br>

- **Ground-effect**: As mentioned previously, due to the ground effect the lift produced will be higher and drag will be lower. This is favourable for take-off, as the required take-off distance will decrease <br><br>

- **Wind**: Under the presence of wind, the aircraft's speed with respect to the ground will be different. To understand how consider the following figure: 

![text](\Images\lec4-wind.png)

The premise of this figure can be described using relative velocity, that is the velocity of the aircraft with respect to the ground can be written in terms of the aircraft's velocity with respect tocthe air and the air's velocity with respect to the ground. The latter term is the speed of wind hence:

$$\{\vec{V_{ac}}\}_{ground} = \{\vec{V_{ac}}\}_{air} + \{\vec{V_{air}}\}_{ground} $$

Therefore, in the presence of headwind, the aircraft's velocity with respect to the ground is decreased, thereby reducing the required take-off distance.

## Landing Maneuver 

The landing maneuver is the trajectory followed by the aircraft from screen height to stand-still. It can be decomposed into two main phases: the airborne and ground run phase. The airborne phase can further be categorised into the approach and flare phases. The ground run can be further divided into de-rotation and deceleration phases. As done for take-off, the total landing distance is the sum of the distances spanned by each of the sub-phases. These two main phases are depicted in the image below:

![image.png](\Images\lec4-landing.png)

The approach phase (of the airborne phase) can be visualised below: 

![text](Images\lec4-landing-traj.png)

For the approach phase, the following equations hold. It is assumed that the velocity is constant and equal to the approach velocity. 
$$ s_{app} = \frac{h_{scr} - h_{fl}}{\tan \gamma_{gs}} $$

$$ h_{fl} = R(1 - \cos \gamma_{gs}) $$

$$ L - W \cos \gamma_{gs} = \frac{W}{g}\frac{V^2}{R} \Rightarrow n -1 = \frac{1}{g}{V^2}{R} $$

Here, we make the aassumption that the load factor during this phase is 1.1 and the approach speed if 1.3 times greater than the minimum, stall speed. This is an emperical approximation that can be used for first-order sizing. 

$$ s_{app} = \frac{h_{scr} - \frac{V_{app}^2}{0.1g}(1 - \cos \gamma_{gs})}{\tan \gamma_{gs}} $$

Secondly, the flare distance is described in the figure below: 

![alt text](Images\lec4-landing-traj2.png)

Based on this figure, the following can be stated:
$$ R = \frac{V_{app}^2}{g \Delta n} $$

$$ s_{fl} = R \sin \gamma_{gs} \Rightarrow s_{fl} = \frac{V_{app}^2}{g \Delta n} \sin \gamma_{gs} $$

While this was not present for take-off, when landing the de-rotation distance must also be considered. This can be calculated as follows:
$$ s_{rot} = V_{app} \Delta t_{rot} = 1.3 V_{min} \Delta t_{rot}$$

Please note the factor of 2.6 is an emperical approximation. Furthermore, to give an order of magnitude, the time for rotation is usually around 2 seconds.

Lastly, the deceleration distance can be computed in a manner similar to the aceleration distance. Let us consider the free body diagram for the deceleration phase: 

![text](\Images\lec4-fbd-landing.png)

Using this free body diagram, the following equations of motion can be noted: 

$$ L - W + N= 0 \\ D + T_{rev} + D_g = m \dot{V} $$

Hence, deceleration can be written as: 

$$ \dot{V} = \frac{W}{g} [T_{rev} + D + \mu_{brake} (W-L)] $$

Now using a fundamental result from kinematics (the same result used during take-off analysis), one can write: 

$$ s_{dec} = \int_0^{x_{dec}} dx = \int_0^{t_{dec}} V dt = \int_{V_{app}}^{0} \frac{V}{a} dV = \frac{1}{\bar{a}} \int_{V_{app}}^{0} VdV = \frac{V_{lof}^2}{2\bar{a}} $$

$$ s_{dec} = 1.3^2 \frac{2}{\rho} \frac{W^2}{S} \frac{1}{C_{L_{max,ld}}} \frac{1}{2g[\bar{T_{rev}} + \bar{D}-\mu_{brake}(W-\bar{L})]} $$

Lastly, the total landing distance is the sum of the distances spanned by the individual phases, and hence:
$$ s_{ld} = s_{app} + s_{fl} + s_{rot} + s_{dec} $$

From this equation, it is evident that to decrease landing distance, either or the following must occur: lower approach speed, decreased weight, increased load factor, or increased glide slope. 

To give a practical consideration: according to regulations the required field length should be a factor 10/6 greater than the demonstrated landing distance.

## Factors influencing Landing Performance

Landing performance, similar to take-off, is dependent on several factors, such as but not limited to:
- **Aerodynamic drag**: Unlike take-off, increasing drag is deterimental for landing and thus sometimes additional measures are used to increase drag. Such an example is the use of spoilers. Spoilers increase aerodynamic drag as well as decrease lift thereby increasing the ground friction as well. <br> <br>

- **Thrust reversers**: In some cases, thrust reversers can be used to temporarely divert the the aircraft engine's thrust such that it acts the direction of motion thereby providing deceleration. <br> <br>

- **Weather and wind**: As described previously under the presence of wind the aircraft's velocity with respect to the ground is different. In the case of headwind, the speed of the aircraft with respect to the ground is decreased and hence the landing distance required is reduced.
